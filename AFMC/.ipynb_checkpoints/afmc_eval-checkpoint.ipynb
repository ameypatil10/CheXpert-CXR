```python
df = pd.read_csv(train_csv).fillna(0).replace({-1: np.NaN}).dropna()
paths = list(df['Path'])
all_studies =  np.unique(np.array(list(map(lambda x: x.split('/view')[0], df['Path']))))
studies = {study: [] for study in all_studies}
# print(len(studies))
# print(np.array(list(map(lambda x: x.split('/view')[0], df['Path']))).shape)

for path in tqdm(paths):
    studies[path.split('/view')[0]].append(path)
    
temp = pd.DataFrame({'Study': all_studies, 'Good': all_studies.shape[0]})
temp.to_csv('all_studies.csv')

idx = 0

good_studies, good_study_idxs = ['example'], [-1]
bad_studies, bad_study_idxs = ['example'], [-1]

# pd.DataFrame({'ID': good_study_idxs, 'Study': good_studies}).to_csv('good_study_idxs.csv', index=False)
# pd.DataFrame({'ID': bad_study_idxs, 'Study': bad_studies}).to_csv('bad_study_idxs.csv', index=False)
```
