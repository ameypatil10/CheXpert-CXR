```python
import pandas as pd
from tqdm import tqdm

data_dir = '/home1/amey/MIMIC/mimic-cxr-jpg-2.0.0.physionet.org/'

df = pd.read_csv(data_dir+'mimic-cxr-2.0.0-metadata.csv.gz')
df
```

```python
split = pd.read_csv('/home1/amey/MIMIC/mimic-cxr-jpg-2.0.0.physionet.org/mimic-cxr-2.0.0-split.csv.gz')
split

data = pd.read_csv('/home1/amey/MIMIC/mimic-cxr-jpg-2.0.0.physionet.org/mimic-cxr-2.0.0-chexpert.csv.gz').fillna(0)
data = data.merge(split)
data

files = ['p{0}/p{1}/s{2}/{3}.jpg'.format(int(data.iloc[idx, 0])//1000000, data.iloc[idx, 0], data.iloc[idx, 1], data.iloc[idx, -2]) for idx in tqdm(range(len(data)))]
data['path'] = files
data = data[['path']+list(data.columns[:-1])]
data.to_csv('data.csv', index=False)
data
```

```python
train_chexpert = data[data['split'] == 'train']
train_chexpert.to_csv('train_chexpert.csv', index=False)
train_chexpert
```

```python
valid_chexpert = data[data['split'] == 'validate']
valid_chexpert.to_csv('valid_chexpert.csv', index=False)
valid_chexpert
```
