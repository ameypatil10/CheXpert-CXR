```python
import os
import cv2
import pandas as pd
import numpy as np
from tqdm import tqdm

train_csv = '/home1/amey/MIMIC/mimic-cxr-jpg-2.0.0.physionet.org/data/mimic-chexpert-train-u1.csv'
data_dir = '/home1/amey/MIMIC/mimic-cxr-jpg-2.0.0.physionet.org/files/'

df = pd.read_csv(train_csv)
df
```

```python
means, stds = [], []
for path in tqdm(df.iloc[:, 0]):
    img = cv2.imread(data_dir+path, 0)/255.0
    means.append(np.mean(img))
    stds.append(np.std(img))

mean, std = np.mean(np.array(means)), np.mean(np.array(stds))
print(mean, std)
```
